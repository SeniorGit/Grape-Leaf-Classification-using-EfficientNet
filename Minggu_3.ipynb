```python
plt.imshow(matriks1,cmap='gray', vmin=0, vmax=255)
```

```python
#Nomor 1 (Flip)
def flip_horizontal(image):
    return [row[::-1] for row in image]

# Function to flip the image vertically
def flip_vertical(image):
    return image[::-1]

# Function to flip the image both horizontally and vertically
def flip_both(image):
    return flip_vertical(flip_horizontal(image))

# Flipping both horizontally and vertically
flipped_image = flip_both(matriks1)

# Displaying the flipped image
plt.imshow(flipped_image,cmap='gray', vmin=0, vmax=255)
```

```python
#Nomor 1 (Flop)
# Function to flop the image (flip horizontally)
def flop(image):
    return [row for row in reversed(image)]

# Flopping the image
flopped_image = flop(matriks1)
plt.imshow(flopped_image,cmap='gray', vmin=0, vmax=255)
```

```python
flipflop = flop(flipped_image)
plt.imshow(flipflop,cmap='gray', vmin=0, vmax=255)
```
